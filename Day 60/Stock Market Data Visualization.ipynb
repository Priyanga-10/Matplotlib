Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Create Small Sample Dataset

```python
data = {
    "Date": pd.date_range(start="2025-01-01", periods=10, freq="B"),  
    "Open": [100, 102, 104, 103, 105, 107, 106, 108, 110, 111],
    "Close": [102, 101, 105, 106, 104, 108, 109, 107, 112, 110],
    "High": [103, 103, 106, 107, 106, 109, 110, 109, 113, 112],
    "Low": [99, 100, 103, 102, 103, 106, 105, 106, 109, 109],
    "Volume": [1500, 1800, 2000, 1700, 1600, 2100, 2200, 1900, 2500, 2300]
}

df = pd.DataFrame(data)
df["Pct_Change"] = ((df["Close"] - df["Open"]) / df["Open"]) * 100
df["Month"] = df["Date"].dt.strftime("%b")
print(df)
```

Line Plot - Stock Prices

```python
plt.figure(figsize=(10,5))
plt.plot(df["Date"], df["Open"], label="Open", linestyle="--", marker="o")
plt.plot(df["Date"], df["Close"], label="Close", linestyle="-", marker="o")
plt.fill_between(df["Date"], df["Low"], df["High"], color="lightblue", alpha=0.3, label="High-Low Range")
plt.title("Daily Stock Price Fluctuations")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.tight_layout()
plt.savefig("line_plot_stock_prices.jpeg", dpi=300)
plt.show()
```

Bar Chart - Trading Volumes

```python
plt.figure(figsize=(10,5))
plt.bar(df["Date"], df["Volume"], color="purple")
plt.title("Daily Trading Volumes")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.tight_layout()
plt.savefig("bar_chart_trading_volumes.jpeg", dpi=300)
plt.show()
```

Scatter Plot - Volume vs % Change

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x="Volume", y="Pct_Change", hue="Pct_Change", palette="coolwarm", size="Pct_Change", sizes=(50,200))
plt.title("Trading Volume vs Percentage Price Change")
plt.xlabel("Trading Volume")
plt.ylabel("% Price Change")
plt.tight_layout()
plt.savefig("scatter_volume_pctchange.jpeg", dpi=300)
plt.show()
```

Box Plot - Monthly Stock Prices

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=df, x="Month", y="Close", palette="Set2")
plt.title("Stock Price Distribution by Month")
plt.xlabel("Month")
plt.ylabel("Closing Price")
plt.tight_layout()
plt.savefig("boxplot_monthly_prices.jpeg", dpi=300)
plt.show()
```

Pairplot - Stock Indicators

```python
sns.pairplot(df[["Open", "Close", "High", "Low", "Volume", "Pct_Change"]], diag_kind="kde")
plt.savefig("pairplot_stock_indicators.jpeg", dpi=300)
plt.show()
```

Heatmap - Correlation

```python
plt.figure(figsize=(7,5))
corr = df[["Open", "Close", "High", "Low", "Volume", "Pct_Change"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Stock Indicators")
plt.tight_layout()
plt.savefig("heatmap_correlation.jpeg", dpi=300)
plt.show()
```
