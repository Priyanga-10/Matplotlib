Import Matplotlib

```python
import matplotlib.pyplot as plt
```

Scatter Plot

```python
# Sample data
experience = [1, 3, 5, 7, 10, 15]
salary = [30000, 40000, 60000, 80000, 120000, 150000]
job_level = ["Junior", "Junior", "Mid", "Mid", "Senior", "Senior"]
departments = ["IT", "HR", "Sales", "Marketing", "IT", "Sales"]

# Colors & sizes for job levels
colors = {"Junior": "blue", "Mid": "green", "Senior": "red"}
sizes = {"Junior": 50, "Mid": 100, "Senior": 200}

# --- Scatter Plot ---
plt.figure(figsize=(10,5))
for i in range(len(experience)):
    plt.scatter(experience[i], salary[i], color=colors[job_level[i]], s=sizes[job_level[i]])

# Add labels for highest paid and lowest experience
plt.text(experience[salary.index(max(salary))], max(salary)+2000, "Highest Paid")
plt.text(min(experience)+0.2, salary[experience.index(min(experience))], "Lowest Exp")

plt.title("Experience vs Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

Pie chart

```python
from collections import Counter
dept_count = Counter(departments)

labels = list(dept_count.keys())
sizes = list(dept_count.values())
explode = [0.1 if s == max(sizes) else 0 for s in sizes]  # highlight largest dept

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90,
        shadow=True, explode=explode)
plt.title("Employee Distribution by Department")
plt.show()
```
