Import Matplotlib

```python
import matplotlib.pyplot as plt
```

Scatter Plot

```python
# Sample Data
ages = [18, 22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
spending = [200, 300, 500, 700, 1000, 850, 600, 400, 300, 250, 150]

# Categorize spending
categories = []
for s in spending:
    if s < 400:
        categories.append("Low")
    elif 400 <= s < 800:
        categories.append("Medium")
    else:
        categories.append("High")

# Assign colors and sizes
colors = {"Low": "skyblue", "Medium": "orange", "High": "red"}
sizes = {"Low": 50, "Medium": 100, "High": 200}

point_colors = [colors[c] for c in categories]
point_sizes = [sizes[c] for c in categories]

# Scatter plot
plt.scatter(ages, spending, c=point_colors, s=point_sizes, edgecolor="black")

# Label high spenders
for age, spend, cat in zip(ages, spending, categories):
    if cat == "High":
        plt.text(age, spend+23, f"${spend}", ha="center", fontsize=8)

plt.title("Customer Age vs Spending")
plt.xlabel("Age")
plt.ylabel("Monthly Spending ($)")
plt.show()
```

Pie Chart

```python
labels = ["Low", "Medium", "High"]
counts = [categories.count(l) for l in labels]
explode = [0.1 if c == max(counts) else 0 for c in counts]

plt.pie(counts, labels=labels, colors=[colors[l] for l in labels],
        autopct="%.1f%%", startangle=90, explode=explode)
plt.title("Spending Category Distribution")
plt.show()
```
