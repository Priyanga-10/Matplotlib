Import Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

 Display Both Charts Together

```python
# Create Sample Sales Data for Scatter Plot
products = ["Laptop", "Phone", "Tablet", "Headphones", "Camera"]
sales_revenue = [50000, 35000, 20000, 15000, 25000]  # in dollars
profit_margin = [12000, 8000, 5000, 2000, 4000]      # in dollars
units_sold = [300, 500, 250, 600, 150]               # for marker sizes

# Create Scatter Plot with Customization
colors = ["green" if profit > 5000 else "red" for profit in profit_margin]
sizes = [u*0.5 for u in units_sold]  # scale marker sizes

# Create Sales Data for Pie Chart
regions = ["North", "South", "East", "West"]
region_sales = [40000, 30000, 20000, 10000]

# Create figure with subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,6))

# ---------------- Scatter Plot ----------------

# Scatter Plot
scatter = ax1.scatter(sales_revenue, profit_margin, 
                      c=colors, s=sizes, alpha=0.7, edgecolors="black")

# Add Labels to Scatter Points
for i, product in enumerate(products):
    ax1.text(sales_revenue[i]+500, profit_margin[i]+300, product, fontsize=9)

# Improve Readability
ax1.set_title("Product Sales vs Profit")
ax1.set_xlabel("Sales Revenue ($)")
ax1.set_ylabel("Profit Margin ($)")
ax1.grid(True, linestyle="--", alpha=0.6)

# ---------------- Pie Chart ----------------

# Customize Pie Chart Colors
colors_pie = ["gold", "skyblue", "lightgreen", "salmon"]

# Highlight Best-Performing Region
explode = [0.1 if sale == max(region_sales) else 0 for sale in region_sales]

# Pie Chart
ax2.pie(region_sales, labels=regions, autopct="%1.1f%%", 
        colors=colors_pie, explode=explode, shadow=True, startangle=90)

ax2.set_title("Sales Contribution by Region")

# Display Both Charts Together
plt.tight_layout()
plt.show()
```
