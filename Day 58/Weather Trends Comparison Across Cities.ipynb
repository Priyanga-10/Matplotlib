Import Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Create Subplots

```python
# Create Sample Data
months = pd.date_range(start="2025-01-01", periods=12, freq="M").strftime("%b")
cities = ["City A", "City B", "City C", "City D"]

np.random.seed(42)
temperature = np.random.randint(15, 35, (12, 4))   # °C
humidity = np.random.randint(40, 90, (12, 4))      # %
rainfall = np.random.randint(50, 300, (12, 4))     # mm
wind_speed = np.random.randint(5, 25, (12, 4))     # km/h

# Create Subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

# --- Temperature Plot ---
for i in range(4):
    axs[0, 0].plot(months, temperature[:, i], marker="o", label=cities[i])
axs[0, 0].set_title("Monthly Temperature (°C)")
axs[0, 0].set_ylabel("Temperature (°C)")
axs[0, 0].legend()

# --- Humidity Plot ---
for i in range(4):
    axs[0, 1].plot(months, humidity[:, i], marker="s", linestyle="--", label=cities[i])
axs[0, 1].set_title("Monthly Humidity (%)")
axs[0, 1].set_ylabel("Humidity (%)")
axs[0, 1].legend()

# --- Rainfall Plot ---
bar_width = 0.2
x = np.arange(len(months))
for i in range(4):
    axs[1, 0].bar(x + i*bar_width, rainfall[:, i], width=bar_width, label=cities[i])
axs[1, 0].set_xticks(x + bar_width*1.5)
axs[1, 0].set_xticklabels(months, rotation=45)
axs[1, 0].set_title("Monthly Rainfall (mm)")
axs[1, 0].set_ylabel("Rainfall (mm)")
axs[1, 0].legend()

# --- Wind Speed Plot ---
for i in range(4):
    axs[1, 1].plot(months, wind_speed[:, i], marker="^", linestyle="-.", label=cities[i])
axs[1, 1].set_title("Monthly Wind Speed (km/h)")
axs[1, 1].set_ylabel("Wind Speed (km/h)")
axs[1, 1].legend()

# Adjust Layout
plt.tight_layout()
plt.show()
```
