Import Required Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Multiple Subplots

```python
# Create Sample Data (7 days × 4 cities)
days = pd.date_range(start="2025-09-01", periods=7, freq="D")
cities = ["City A", "City B", "City C", "City D"]

np.random.seed(58)  # for reproducibility
PM25 = np.random.randint(50, 150, (7, 4))   # PM2.5 levels
PM10 = np.random.randint(80, 200, (7, 4))   # PM10 levels
CO   = np.random.uniform(0.5, 2.5, (7, 4))  # CO concentration
NO2  = np.random.randint(20, 80, (7, 4))    # NO2 concentration

# Initialize a Figure with Subplots (2 × 2 grid)
fig, axs = plt.subplots(2, 2, figsize=(14, 10), sharex=True)

# Plot PM2.5 Levels (Line chart for each city)
for i in range(4):
    axs[0, 0].plot(days, PM25[:, i], marker="o", label=cities[i])
axs[0, 0].set_title("PM2.5 Levels Across Cities")
axs[0, 0].set_ylabel("PM2.5 (µg/m³)")
axs[0, 0].grid(True, linestyle="--", alpha=0.6)
axs[0, 0].legend()

# Plot PM10 Levels (Bar chart for each city, grouped by day)
bar_width = 0.5
x = np.arange(len(days))
for i in range(4):
    axs[0, 1].bar(x + i*bar_width, PM10[:, i], width=bar_width, label=cities[i])
axs[0, 1].set_title("PM10 Levels Across Cities")
axs[0, 1].set_ylabel("PM10 (µg/m³)")
axs[0, 1].set_xticks(x + bar_width*1.5)
axs[0, 1].set_xticklabels(days.strftime("%b %d"), rotation=45)
axs[0, 1].grid(axis="y", linestyle="--", alpha=0.6)
axs[0, 1].legend()

# Plot CO Concentration (Scatter plot with color variation per city)
colors = ["red", "blue", "green", "orange"]
for i in range(4):
    axs[1, 0].scatter(days, CO[:, i], color=colors[i], label=cities[i], s=80, alpha=0.7)
axs[1, 0].set_title("CO Concentration Across Cities")
axs[1, 0].set_ylabel("CO (mg/m³)")
axs[1, 0].grid(True, linestyle="--", alpha=0.6)
axs[1, 0].legend()

# Plot NO2 Concentration (Histogram of frequency distribution)
axs[1, 1].hist(NO2.flatten(), bins=10, color="purple", edgecolor="black", alpha=0.7)
axs[1, 1].set_title("NO2 Concentration Distribution")
axs[1, 1].set_xlabel("NO2 (µg/m³)")
axs[1, 1].set_ylabel("Frequency")
axs[1, 1].grid(axis="y", linestyle="--", alpha=0.6)

plt.tight_layout()

plt.savefig("Air_Pollution_Trends.png", dpi=300)

plt.show()
```
