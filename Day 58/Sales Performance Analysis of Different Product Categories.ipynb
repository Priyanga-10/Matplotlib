Import Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Create Sample Data

```python
# 12 months
months = np.arange(1, 13)

# Random sales data for 4 product categories
np.random.seed(42)
electronics = np.random.randint(200, 500, size=12)
clothing = np.random.randint(150, 400, size=12)
home_appliances = np.random.randint(100, 300, size=12)
books = np.random.randint(80, 200, size=12)
```

Create Subplots

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Electronics
axs[0, 0].plot(months, electronics, marker='o', color='blue')
axs[0, 0].set_title("Electronics Sales")
axs[0, 0].set_ylabel("Sales ($)")
axs[0, 0].grid(True, linestyle="--", alpha=0.6)

# Clothing
axs[0, 1].plot(months, clothing, marker='s', color='green')
axs[0, 1].set_title("Clothing Sales")
axs[0, 1].grid(True, linestyle="--", alpha=0.6)

# Home Appliances
axs[1, 0].plot(months, home_appliances, marker='^', color='orange')
axs[1, 0].set_title("Home Appliances Sales")
axs[1, 0].set_xlabel("Months")
axs[1, 0].set_ylabel("Sales ($)")
axs[1, 0].grid(True, linestyle="--", alpha=0.6)

# Books
axs[1, 1].plot(months, books, marker='d', color='red')
axs[1, 1].set_title("Books Sales")
axs[1, 1].set_xlabel("Months")
axs[1, 1].grid(True, linestyle="--", alpha=0.6)

# Adjust layout
plt.tight_layout()
plt.show()
```
