Import Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Sample Dataset

```python
data = {
    "Employee_ID": ["E001", "E002", "E003", "E004", "E005", "E006", "E007", "E008"],
    "Department": ["IT", "HR", "Finance", "IT", "Finance", "HR", "IT", "Finance"],
    "Salary": [75000, 50000, 65000, 80000, 70000, 52000, 90000, 72000]
}

df = pd.DataFrame(data)
print(df)
```

Department-Wise Average Salary (Bar Chart)

```python
# Group by department and calculate average salary
dept_avg = df.groupby("Department")["Salary"].mean()

# Bar chart
plt.figure(figsize=(7,5))
dept_avg.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.xticks(rotation=0)
plt.show()
```

Overall Salary Distribution (Histogram)

```python
plt.figure(figsize=(7,5))
plt.hist(df["Salary"], bins=5, color="lightgreen", edgecolor="black")
plt.title("Salary Distribution of Employees")
plt.xlabel("Salary")
plt.ylabel("Number of Employees")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
