Import Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

1. Bar Chart for Movie Genres

```python
# Sample movie genres and earnings (in millions)
genres = ["Action", "Comedy", "Drama", "Horror", "Sci-Fi"]
earnings = [850, 600, 400, 300, 700]

# Bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(genres, earnings, color=["skyblue", "lightgreen", "salmon", "orange", "violet"], width=0.6)

# Labels and title
plt.xlabel("Movie Genre")
plt.ylabel("Total Earnings (in Millions)")
plt.title("Total Box Office Earnings by Genre")

# Display values on top of bars
for i in range (len(genres)):
    plt.text(i, earnings[i] + 20, str(earnings[i]), ha = 'center', fontsize = 10)

# Save the bar chart
plt.savefig("genre_earnings_bar_chart.png")
plt.show()
```

2. Horizontal Bar Chart

```python
genres = ["Action", "Comedy", "Drama", "Horror", "Sci-Fi"]
earnings = [850, 600, 400, 300, 700]

plt.figure(figsize=(8, 6))
bars_h = plt.barh(genres, earnings, color="lightcoral", height=0.5)

# Labels and title
plt.xlabel("Total Earnings (in Millions)")
plt.ylabel("Movie Genre")
plt.title("Total Box Office Earnings by Genre (Horizontal)")

# Display values on bars
for i in range (len(genres)):
    plt.text(earnings[i] + 20, i, str(earnings[i]), va = 'center', fontsize = 10)

# Save the horizontal bar chart
plt.savefig("genre_earnings_horizontal_bar_chart.png")
plt.show()
```

3. Histogram for Movie Earnings

```python
# Generate random earnings for 100 movies (in millions)
np.random.seed(42)
movie_earnings = np.random.randint(50, 1000, size=100)

plt.figure(figsize=(8, 6))
plt.hist(movie_earnings, bins=10, color="skyblue", edgecolor="black")

# Labels and title
plt.xlabel("Earnings (in Millions)")
plt.ylabel("Number of Movies")
plt.title("Distribution of Movie Earnings")

# Grid for readability
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Save the histogram
plt.savefig("movie_earnings_histogram.png")
plt.show()
```
