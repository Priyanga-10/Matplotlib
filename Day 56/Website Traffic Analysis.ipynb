Import Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Create DataFrame

```python
data = {
    "Date": ["2025-09-01", "2025-09-01", "2025-09-01"],
    "Page": ["Homepage", "Blog", "Contact"],
    "Visitors": [1200, 800, 300],
    "Session_Duration": [5.2, 3.4, 2.1]
}

df = pd.DataFrame(data)
print(df)
```

Bar Chart: Daily Visitors per Page

```python
# Aggregate visitors by page
page_visitors = df.groupby("Page")["Visitors"].sum()

# Bar chart
plt.figure(figsize=(8,6))
plt.bar(page_visitors.index, page_visitors.values, color="skyblue", edgecolor="black")

plt.title("Total Visitors per Page")
plt.xlabel("Website Pages")
plt.ylabel("Total Visitors")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

Histogram: Session Duration Distribution

```python
plt.figure(figsize=(8,6))
plt.hist(df["Session_Duration"], bins=10, color="lightgreen", edgecolor="black", alpha=0.7)

plt.title("Distribution of Session Durations")
plt.xlabel("Session Duration (minutes)")
plt.ylabel("Frequency")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
