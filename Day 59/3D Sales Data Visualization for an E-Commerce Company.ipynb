Import Required Libraries

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

Sample E-commerce Data

```python
categories = ["Electronics", "Clothing", "Home", "Books", "Sports"]
regions = ["North", "South", "East", "West"]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]

np.random.seed(42)
revenue = np.random.randint(5000, 20000, size=len(categories))
customers = np.random.randint(100, 1000, size=len(categories))
orders = np.random.randint(50, 500, size=len(categories))

monthly_sales = np.random.randint(10000, 40000, size=(len(months), len(regions)))
```

3D Scatter Plot

```python
fig = plt.figure(figsize=(7,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(revenue, customers, orders, c=orders, cmap="viridis", s=80)

ax.set_xlabel("Revenue")
ax.set_ylabel("Customers")
ax.set_zlabel("Orders")
ax.set_title("3D Scatter: Sales Trends by Category")

# Annotate best-selling category
best_idx = np.argmax(revenue)
ax.text(revenue[best_idx], customers[best_idx], orders[best_idx],
        categories[best_idx], color="red", fontsize=10)

plt.show()
```

3D Bar Chart

```python
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

dx = dy = 0.4
z = 0  # base level

# Loop through months & regions
for i, month in enumerate(months):
    for j, region in enumerate(regions):
        ax.bar3d(i, j, z, dx, dy, monthly_sales[i, j], shade=True)

# Axis labels
ax.set_xticks(np.arange(len(months)) + dx/2)
ax.set_xticklabels(months)
ax.set_yticks(np.arange(len(regions)) + dy/2)
ax.set_yticklabels(regions)
ax.set_zlabel("Sales")
ax.set_title("3D Bar Chart: Monthly Sales by Region")

plt.show()
```

Heatmap of Sales Hours

```python
hours = np.arange(24)
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
sales_heatmap = np.random.randint(50, 500, size=(len(days), len(hours)))

plt.figure(figsize=(12,5))
plt.imshow(sales_heatmap, cmap="YlGnBu", aspect="auto")

# Color bar
plt.colorbar(label="Sales")

# Axis labels
plt.xticks(ticks=np.arange(len(hours)), labels=hours)
plt.yticks(ticks=np.arange(len(days)), labels=days)
plt.title("Heatmap: Peak Sales by Hour & Day")
plt.xlabel("Hour of Day")
plt.ylabel("Day of Week")

plt.show()
```
