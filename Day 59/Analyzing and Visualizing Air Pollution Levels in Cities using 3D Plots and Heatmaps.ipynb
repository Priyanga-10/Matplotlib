 Import Required Libraries

```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D 
```

 Prepare Air Pollution Data

```python
# We consider four cities and three pollutants (PM2.5, NO2, CO). 
# City names 
cities = ["New York", "Los Angeles", "Chicago", "Houston"] 
 
# Pollution Data (PM2.5, NO2, CO in µg/m³) 
pollution_levels = np.array([ 
    [35, 50, 7],  # New York 
    [55, 70, 10], # Los Angeles 
    [40, 60, 8],  # Chicago 
    [30, 45, 6]   # Houston 
]) 
 
# Separate pollution levels 
pm25 = pollution_levels[:, 0] 
no2 = pollution_levels[:, 1] 
co = pollution_levels[:, 2] 
```

Create a 3D Scatter Plot for Pollution Levels 

```python
# Create figure 
fig = plt.figure(figsize=(8, 6)) 
ax = fig.add_subplot(111, projection='3d') 
 
# Scatter plot 
ax.scatter(pm25, no2, co, color='red', marker='o') 
 
# Labels 
ax.set_xlabel("PM2.5 (µg/m³)") 
ax.set_ylabel("NO2 (µg/m³)") 
ax.set_zlabel("CO (µg/m³)") 
ax.set_title("3D Scatter Plot of Air Pollution Levels") 
 
# Annotate most polluted city 
ax.text(55, 70, 10, "Los Angeles", color='blue', fontsize=10) 
 
# Show plot 
plt.show() 
```

 Create a 3D Bar Chart to Compare Cities 

```python
fig = plt.figure(figsize=(8, 6)) 
ax = fig.add_subplot(111, projection='3d') 
# X-axis positions 
x_pos = np.arange(len(cities)) 
y_pos = np.zeros_like(x_pos)   
 
# Bar chart for each pollutant 
ax.bar3d(x_pos, y_pos, np.zeros_like(pm25), dx=0.5, dy=0.5, dz=pm25, color='r', label="PM2.5") 
ax.bar3d(x_pos, y_pos+1, np.zeros_like(no2), dx=0.5, dy=0.5, dz=no2, color='g', label="NO2") 
ax.bar3d(x_pos, y_pos+2, np.zeros_like(co), dx=0.5, dy=0.5, dz=co, color='b', label="CO") 
 
# Labels 
ax.set_xticks(x_pos) 
ax.set_xticklabels(cities) 
ax.set_xlabel("Cities") 
ax.set_ylabel("Pollutants") 
ax.set_zlabel("Concentration (µg/m³)") 
ax.set_title("3D Bar Chart of Air Pollution") 
 
# Show plot 
plt.legend() 
plt.show()
```

Create a Heatmap of Monthly Pollution Trends

```python
# Monthly pollution data (rows: months, columns: cities) 
monthly_pollution = np.random.randint(20, 80, size=(12, 4)) 
 
# Create Heatmap 
plt.figure(figsize=(8, 5)) 
plt.imshow(monthly_pollution, cmap='coolwarm', interpolation='nearest') 
 
# Add color bar 
plt.colorbar(label="Pollution Level (µg/m³)") 
 
# Labels 
plt.xticks(ticks=np.arange(len(cities)), labels=cities, rotation = 45, ha = "right") 
plt.yticks(ticks=np.arange(12), labels=["Jan", "Feb", "Mar", "Apr", "May", "Jun", 
"Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]) 
plt.title("Heatmap of Monthly Pollution Trends") 
 
# Show plot 
plt.show()
```
