Import Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

Load & Prepare Weather Data

```python
cities = ["City A", "City B", "City C", "City D"]
days = np.arange(1, 8)   # 7 days
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

np.random.seed(42)
temp = np.random.randint(15, 40, (4, 7))     # Temperature
hum = np.random.randint(40, 90, (4, 7))      # Humidity
wind = np.random.randint(5, 20, (4, 7))      # Wind Speed
avg_temp = np.random.randint(15, 35, (12, 4))# Avg Monthly Temp
avg_hum = np.random.randint(40, 90, (12, 4)) # Avg Monthly Humidity
```

 3D Scatter Plot

```python
fig = plt.figure(figsize=(8,7))
ax = fig.add_subplot(111, projection="3d")
for i, city in enumerate(cities):
    sc = ax.scatter(temp[i], hum[i], wind[i], 
                    c=temp[i], cmap="coolwarm", 
                    s=temp[i]*5, label=city)
ax.set_xlabel("Temperature (C)")
ax.set_ylabel("Humidity (%)")
ax.set_zlabel("Wind Speed (km/h)")
ax.set_title("3D Scatter Plot of Weather Data")
ax.legend()
```

3D Line Plot

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="3d")
for i, city in enumerate(cities):
    ax.plot(days, temp[i], hum[i], label=city)
ax.set_xlabel("Days")
ax.set_ylabel("Temperature (C)")
ax.set_zlabel("Humidity (%)")
ax.set_title("3D Line Plot - Weekly Trend")
ax.legend()
```

3D Bar Chart (Monthly Avg Temp)

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection="3d")
_x = np.arange(len(cities))
_y = np.arange(12)
_xx, _yy = np.meshgrid(_x, _y)
x, y = _xx.ravel(), _yy.ravel()
z = np.zeros_like(x)
dz = avg_temp.ravel()
ax.bar3d(x, y, z, 0.5, 0.5, dz, shade=True)
ax.set_xlabel("Cities")
ax.set_ylabel("Months")
ax.set_zlabel("Avg Temp (C)")
ax.set_title("3D Bar Chart - Monthly Avg Temp")
```

 Heatmap (Temperature)

```python
plt.figure(figsize=(8,6))
plt.imshow(avg_temp,  aspect="auto")
plt.colorbar(label="Temperature (C)")
plt.xticks(np.arange(len(cities)), cities)
plt.yticks(np.arange(12), months)
plt.title("Heatmap - Temperature Variation Across Months")
plt.show()
```

Compare Temp & Humidity Heatmaps

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))
im1 = ax1.imshow(avg_temp, cmap="coolwarm", aspect="auto")
ax1.set_title("Temperature Heatmap")
ax1.set_xticks(np.arange(len(cities)), labels=cities)
ax1.set_yticks(np.arange(12), labels=months)
fig.colorbar(im1, ax=ax1)

im2 = ax2.imshow(avg_hum, cmap="Blues", aspect="auto")
ax2.set_title("Humidity Heatmap")
ax2.set_xticks(np.arange(len(cities)), labels=cities)
ax2.set_yticks(np.arange(12), labels=months)
fig.colorbar(im2, ax=ax2)

plt.tight_layout()
plt.show()
```

Interactive 3D Wind Speed Plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
for i, city in enumerate(cities):
    ax.scatter(temp[i], hum[i], wind[i], 
               c=wind[i], cmap="viridis", 
               s=wind[i]*10, marker="o", label=city)
ax.set_xlabel("Temperature")
ax.set_ylabel("Humidity")
ax.set_zlabel("Wind Speed")
ax.set_title("Interactive 3D Wind Speed Plot")
ax.legend()
plt.show()
```

Dashboard (Quick Layout)

```python
fig = plt.figure(figsize=(12,10))

# 3D Scatter
ax1 = fig.add_subplot(221, projection="3d")
ax1.scatter(temp[0], hum[0], wind[0], c=temp[0], cmap="coolwarm", s=60)
ax1.set_title("3D Scatter")

# 3D Line
ax2 = fig.add_subplot(222, projection="3d")
for i in range(len(cities)):
    ax2.plot(days, temp[i], hum[i])
ax2.set_title("3D Line")

# 3D Bar
ax3 = fig.add_subplot(223, projection="3d")
ax3.bar3d(x, y, z, 0.5, 0.5, dz)
ax3.set_title("3D Bar Chart")

# Heatmap
ax4 = fig.add_subplot(224)
im = ax4.imshow(avg_temp, cmap="plasma", aspect="auto")
ax4.set_title("Temperature Heatmap")
plt.colorbar(im, ax=ax4)

plt.tight_layout()
plt.show()
```
